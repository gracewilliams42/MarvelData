```python
# what if I wanted to split the files into a list of words?
thorDialogue=[]
with open ("Thor.txt") as thor:
  lines = thor.readlines()
  for line in lines:
      words = line.split()
      for word in words:
          thorDialogue.append(word)
        
print(thorDialogue)
```

```python
# create a dictionary of all the words, then make the values for each word the # of occurences 
# to look at what words occur the most often
thorDialogueDict={}
for word in thorDialogue:
    if word in thorDialogueDict:
        thorDialogueDict[word]+=1
    else: 
        thorDialogueDict[word]=1
```

```python
thorDialogueDict
```

```python
# input the sortDictionaryValues function from sortdictionary.py
# to sort the dict of words by value
def sortDictionaryByValues(dictionaryName, ascending=True):
    if (type(ascending) != bool and type(dictionaryName) != dict):
        raise ValueError("sortDictionaryByValues requires a dictionary and boolean as arguments, respectively")
        return None
    else:
        sortedDictionary = {}
        sortedList = sorted(dictionaryName.items(),key=lambda x: x[1], reverse=(not ascending))
        for item in sortedList:
            sortedDictionary[item[0]] = item[1]
        return sortedDictionary

sortDictionaryByValues(thorDialogueDict, ascending=False)
```

```python
# What if I went through and made a list of all of the titles,
#then I did a function to open each one and record the length in a dictionary to turn 
# into a dataframe
def lengthOfMovieDialogues(listOfFiles): 
    dictOfTitlesAndLengths={}
    for file in listOfFiles: 
        with open(file,"r",encoding="utf8") as fileName:
            fileOpen = fileName.read()
            fileNameForDict=file[0:-3]
            dictOfTitlesAndLengths[fileNameForDict]=len(fileOpen)
    return(dictOfTitlesAndLengths)
    
```

```python
AvengersFiles=["Avengers.Age.of.Ultron.txt","Avengers.Infinity.War.txt","Avengers.Endgame.txt"]
lengthOfMovieDialogues(AvengersFiles)
```

```python
# now let's do it with all of the movie txt files 
#except for avengers, ant man and the wasp and Black Panther and Cap the first avenger 
# and Spiderman homecoming because those ones are unreadable? 
allFilmTitles=["Thor.txt","Thor.The.Dark.World.txt","Thor.Ragnarok.txt","Spider-Man.Far.From.Home.txt","Iron-Man.txt","Iron-Man.2.txt","Iron-Man.3.txt","Guardians.of.the.Galaxy.txt","Guardians.of.the.Galaxy.Vol. 2.txt","Doctor.Strange.txt","Captain.Marvel.txt","Captain.America.The.Winter.Soldier.txt","Captain.America.Civil.War.txt","Avengers.Age.of.Ultron.txt","Avengers.Infinity.War.txt","Avengers.Endgame.txt","Ant-Man.txt"]
```

```python
len(allFilmTitles) # 17 films in the dictionary 
```

```python
# let's look at the dictionary 
lengthOfMovieDialogues(allFilmTitles)
```

```python
dictOfTitlesAndLengths1={}
for file in allFilmTitles: 
    with open(file,"r",encoding="utf8") as fileName:
        fileOpen = fileName.read()
        fileNameForDict=file[0:-3]
        dictOfTitlesAndLengths1[fileNameForDict]=len(fileOpen)
# turn the dictionary into a dataframe
filmLengthsDF = pd.DataFrame(list(dictOfTitlesAndLengths1.items()),columns = ['Title','File Length']) 
```

```python
# sort the dataframe by file length
FilmLengthsToPlot=filmLengthsDF.sort_values("File Length", ascending=False)

FilmLengthsToPlot
# this says that the file for Iron-Man 3 is longest, and Avengers Endgame is third?
# This may make sense, since the avengers movies, while they are longer, usually 
# have longer fight scenes and less dialogue 
```

```python
# let's make it into a bar chart to display all the titles and their number of words 
import matplotlib.pyplot as plt
plt.bar(FilmLengthsToPlot["Title"],FilmLengthsToPlot["File Length"])
plt.title('Number of Words in Marvel Movies')
plt.xlabel('Title')
plt.ylabel('File Length')
plt.xticks(rotation=90)
plt.savefig('NumWordsByMovie.jpg')

plt.show()
```

```python
# lets add another column that includes the year of each movie's release 
listOfReleaseYears=[2013,2015,2019,2010,2019,2016,2008,2015,2017,2017,2018,2014,2014,2019,2016,2011,2013]
FilmLengthsToPlot.insert(2, "Release Year", listOfReleaseYears, True) 
```

```python
FilmLengthsToPlot
```

```python
# group the dataframe by year and aggregate to get the sum of all of the words said in 
# movies that came out each year
FilmLengthsGrouped=FilmLengthsToPlot.groupby(FilmLengthsToPlot["Release Year"]).agg({"File Length":np.mean})
```

```python
FilmLengthsGrouped
# reset index to make sure all of the columns are right 
FilmLengthsGroupedNew=FilmLengthsGrouped.reset_index()
FilmLengthsGroupedNew
```

```python
# let's plot the total number of words said in marvel movies each year
sb.catplot(x=FilmLengthsToPlot["Release Year"],y=FilmLengthsToPlot["File Length"], data=FilmLengthsGroupedNew,height=10,kind="bar")
```

```python
# let's plot the total number of words said in marvel movies each year
import matplotlib.pyplot as plt
FilmLengthsGroupedNewToPlot=FilmLengthsGroupedNew.sort_values
plt.bar(FilmLengthsGroupedNew["Release Year"],FilmLengthsGroupedNew["File Length"])
plt.title('Average Number of Words in Marvel Movies by Year')
plt.xlabel('Year')
plt.ylabel('Average Number of Words in Movie Scripts')
plt.xticks(rotation=90)
plt.savefig('AvgNumWords.png')
plt.show()
```
